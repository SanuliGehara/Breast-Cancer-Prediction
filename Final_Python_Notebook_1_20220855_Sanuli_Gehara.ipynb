Below code block is reused from the code reuse session 1 to discover the outliers.

This method(function) finds and returns the outliers for a column using IQR

```python
#Function to discover outliers using IQR
def find_outliers_IQR(data):
  q1=data.quantile(0.25)
  q3=data.quantile(0.75)
  IQR=q3-q1

  outliers = data[((data<(q1-1.5*IQR))|(data> (q3+1.5*IQR)))]

  return outliers
```

```python
# Find and Print the outliers for Age
outliers = find_outliers_IQR(data_frame['Age'])

print("number of outliers: "+ str(len(outliers)))

print(outliers)
```
